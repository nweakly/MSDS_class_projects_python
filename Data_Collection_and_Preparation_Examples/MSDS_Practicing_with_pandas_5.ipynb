```python
#load data from the file
clients=pd.read_csv('bank-additional.csv', sep=';')   
```

```python
#Look at the structure of the dataframe
clients.info()
```

```python
#change column names
#strip potential extra white spaces
#replace '.' and '_'   with spaces, capitalize
clients.columns = clients.columns.str.strip().str.replace('.', ' ').str.replace('_', ' ').str.capitalize()
```

```python
#Rename some detaframe columns to give them more descriptive names
clients.rename(columns={'Job': 'Job Type', 'Marital': 'Marital Status', 'Default': 'Is In Default', \
                        'Pdays': 'Days Since Last Contact', 'Previous': 'Previous Contacts',  \
                        'Poutcome': 'Past Outcome', 'Emp var rate': 'Employment Variation Rate', \
                       'Cons price idx': 'Consumer Price Index', 'Cond conf idx': 'Consumer Confidence Index', \
                       'Euribor3m': 'Euribor 3 Month Rate', 'Nr employed': 'Number of Employees', \
                       'Y': 'Has Term Deposit'}, inplace=True)
```

```python
#Check results
clients.info()
```

```python
#Change column types
#convert columns to 'category' type 
columnsToCategories=['Job Type', 'Marital Status', 'Education', 'Is In Default', 'Housing', "Loan", 'Contact', 'Month', 'Day of week', 'Past Outcome']
for i in columnsToCategories:
    clients[i]= clients[i].astype('category')
```

```python
#Convert 'Has Term Deposit' to boolean
clients['Has Term Deposit'] = clients['Has Term Deposit'].astype('bool')
```

```python
#Check results
clients.info()
```

```python
#Replace '-' and '.' in strings with spaces in the 'Job Type' column
#get rid of trailing  while spaces after processing 'admin.' category
clients['Job Type'] = clients['Job Type'].str.replace('.', ' ').str.strip().astype('category')
clients['Job Type'] = clients['Job Type'].str.replace('-', ' ').astype('category')
```

```python
#check results
clients['Job Type']
```

```python
#Replace '.' in strings with spaces in the 'Education' column
clients['Education'] = clients['Education'].str.replace('.', ' ').astype('category')
```

```python
#check results
clients['Education']
```

```python
# Use groupby() to calculate how many various levels exist in "Days Since Last Contact"
# and how many clients belong in each group. 
clients.groupby('Days Since Last Contact').count()
```

```python
#The "Days Since Last Contact" column contains 3959 values = 999 (out of 4119 total observations) to indicate that the client has not been contacted before
#Deleting or imputing these values will introduce a significant bias in the data set. 
#Depending on the task it might make be useful to keep these values as '999', however for many tasks (e.g., calculating descriptive statistics
# with automatically omnitting non-existent values) it might be better to replace them with NaNs.

#Replace 999 with NaN

clients['Days Since Last Contact'] = clients['Days Since Last Contact'].replace(999, np.NaN)
```

```python
#check results
clients.info()
```

```python
# Dealing with "unknown" values
# When converting string columns to 'category' data type, "unknown" values present in  
# six columns ('Job Type',  'Marital Status', 'Education', 'Is In Default', 'Housing', and 'Loan'),
# were automatically converted to separate categories - see the output below. 

print(columnsToCategories)
for i in columnsToCategories:
    print(clients[i].cat.categories)
    
```
