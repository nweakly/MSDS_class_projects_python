### Join and Merge

```python
#Left join - employees and departments tables
pd.merge(employees, departments, on='departmentId', how='left')
```

The above output includes all records from the employees table (including past/newly hired unassigned employees), but only those departments that have employees assigned to them (e.g., there is no manufacturing department on the list).

```python
#Right join - employees and branchoffices table
pd.merge(employees, branchoffices, on='branchId', how='right')
```

The output above includes all branch offices from the branchoffices table (including branch with the Id 'b4' -"Middle of Nowhere" that does not have any employees assigned to it), but only current employees (e.g. employees without a branch assignment are excluded).  

```python
#Outer join
pd.merge(employees, departments, on='departmentId', how='outer')
```

The outer join produced the output that includes all records from the employees table and all records from the departments table (including unassigned employees and the Manufacturing department that currently does not have any employees in it).

```python
#Inner join
pd.merge(employees, departments, how='inner')
```

The inner join resulted in a set of records that have matches in both tables (all current employees assigned to departments and only departments that have employees).

```python
#Show a list of current employees, their corresponding departments and  branch offices
pd.merge(pd.merge(employees, departments, how='inner'), branchoffices, how='inner')
```

The above output shows all current employees and their corresponding departments and branch offices.

```python
#Alternative solution with additional sort by department, then by branch
employees.merge(departments, how='inner').merge(branchoffices, how='inner').sort_values(by=['departmentId', 'branchId'])
```

```python
#Display full information only for employees assigned to a particular department (e.g., IT) using a boolean variable
deptIt= employees['departmentId'] == 'd1'
employees[deptIt].merge(departments, how='inner').merge(branchoffices, how='inner').sort_values(by=['departmentId', 'branchId'])
```
