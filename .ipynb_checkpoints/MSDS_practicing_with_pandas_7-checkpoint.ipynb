### Deleting unnecessary columns

```python
#columns for monthly summary data are empty => Delete  
weather.drop(weather.columns[48:], axis=1, inplace=True)
```

In the previous step, 45 rows with daily summaries and no weather observations were deleted. As a result, the 'DAILYWeather' column and the last 8 columns are empty => Delete

```python
#Check results
weather.info()
```

```python
# Drop column with the weather station info - it is the same for all data points
weather.drop(['STATION', 'STATION_NAME', 'ELEVATION', 'LONGITUDE', 'LATITUDE'], axis=1, inplace=True)
#Drop other empty columns
weather.drop(['DAILYAverageRelativeHumidity', 'DAILYAverageDewPointTemp', 'DAILYAverageWetBulbTemp', 'DAILYAverageSeaLevelPressure'], axis=1, inplace=True)
```

```python
#check results
weather.info()
```

```python
# Drop columns that contain daily summary info 
weather.drop(weather.columns[29:], axis=1, inplace=True)
weather.drop(['DAILYMaximumDryBulbTemp', 'DAILYMinimumDryBulbTemp', 'DAILYAverageDryBulbTemp', 'DAILYDeptFromNormalAverageTemp', 'DAILYHeatingDegreeDays', 'DAILYCoolingDegreeDays'], axis=1, inplace=True)
```

### Fortmatting

```python
#Rename the dataframe columns to give them more descriptive names
weather.rename(columns={'DATE': 'Date', 'REPORTTPYE': 'Report Type', 'HOURLYSKYCONDITIONS': 'Sky Conditions', \
                        'HOURLYVISIBILITY': 'Visibility', 'HOURLYPRSENTWEATHERTYPE': 'Present Weather Type',  \
                        'HOURLYDRYBULBTEMPF': 'Dry Bulb Temperature F', 'HOURLYDRYBULBTEMPC': 'Dry Bulb Temperature C', \
                       'HOURLYWETBULBTEMPF': 'Wet Bulb Temperature F', 'HOURLYWETBULBTEMPC': 'Wet Bulb Temperature C', \
                       'HOURLYDewPointTempF': 'Dew Point F', 'HOURLYDewPointTempC': 'Dew Point C', \
                       'HOURLYRelativeHumidity': 'Relative Humidity', 'HOURLYWindSpeed': 'Wind Speed', \
                       'HOURLYWindDirection': 'Wind Direction', 'HOURLYWindGustSpeed': 'Wind Gust Speed', \
                       'HOURLYStationPressure': 'Station Pressure', 'HOURLYPressureTendency': 'Pressure Tendency', \
                       'HOURLYPressureChange': 'Pressure Change', 'HOURLYSeaLevelPressure': 'Sea Level Pressure', \
                       'HOURLYPrecip': 'Precipitation', 'HOURLYAltimeterSetting': 'Altimeter Setting', \
                       'DAILYSunrise': 'Sunrise', 'DAILYSunset': 'Sunset'}, inplace=True)
```

```python
#Check results
weather.info()
```

```python
#Convert 'Dry Bulb Temperature F', 'Dry Bulb Temperature C'  and 'Wind Speed' to numeric
weather['Dry Bulb Temperature F']=pd.to_numeric(weather['Dry Bulb Temperature F'], errors='coerce')
weather['Dry Bulb Temperature C']=pd.to_numeric(weather['Dry Bulb Temperature C'], errors='coerce')
weather['Wind Speed']=pd.to_numeric(weather['Wind Speed'], errors='coerce')
```

```python
#Fill in NaNs created by the data conversions with the mean values
weather['Dry Bulb Temperature F'].fillna(weather['Dry Bulb Temperature F'].mean(), inplace=True)
weather['Dry Bulb Temperature C'].fillna(weather['Dry Bulb Temperature C'].mean(), inplace=True)
weather['Wind Speed'].fillna(weather['Wind Speed'].mean(), inplace=True)
```

```python
weather.info()
```

```python
#Missing values
#Drop the "Present Weather Type" column - not enough observation to impute missing values 
weather.drop('Present Weather Type', axis=1, inplace=True)
```

```python
# fill in NaNs with '0' for the "Wind Gust Speed",  "Pressure Tendency", "Pressure change" 
# if observations were not recorded, then there were no changes in the related columns (wind speed, pressure)
col_zeros = ['Wind Gust Speed', 'Pressure Tendency', 'Pressure Change']
weather[col_zeros] = weather[col_zeros].apply(lambda x: x.fillna(0))
```

```python
#Fill in NaNs in the "Altimeter Setting" column with the mean
weather['Altimeter Setting'].fillna(weather['Altimeter Setting'].mean(), inplace=True)
```

```python
# Keep 'T' - for traces in the 'Precipitation' column,  fill NaNs with 0
weather['Precipitation'] = weather['Precipitation'].fillna(0)
```

```python
# use interpolation to fill in missing values in the 'Visibility' and 'Sea Level Pressure' columns
# as observations for continious processes are likely to be related 
weather['Visibility'].interpolate(method='pad', inplace=True)
weather['Sea Level Pressure'].interpolate(method='pad', inplace=True)
weather['Sea Level Pressure'].fillna(method='bfill', inplace=True) #first two observations
```

```python
# Use "Not Recorded" instead of NaN in the 'Sky Conditions' column  since there is no suitable values to impute missing observations
weather['Sky Conditions'].fillna('Not Recorded', inplace=True)
```

```python
#Check results
weather.info()
```

```python
#look at the remaining missing observations (e.g., in the "Wind Direction" column)
null_data = weather[weather.isnull().any(axis=1)]
null_data.iloc[:, 5:]
```

```python
#look at the same day weather data
weather['2019-01-09'].iloc[:, 5:]
```

```python
weather['2019-02-02'].iloc[:, 5:]
```

```python
# There are two missing observations in the "Wind Direction" column, both are less than 10 minutes 
# after a previous observation with 'VRB' for variable wind direction =>
# fill in missing values with previous values 
weather['Wind Direction'].fillna(method='ffill', inplace=True)       
```

```python
#There is no more missing values in the data frame - it is ready for the analysis
weather.info()
```

### Analysis
#### Temperatures

```python
# Maximum Dry Bulb temperature observed in January (in degrees Celsius)
weather['2019-01']['Dry Bulb Temperature C'].max()
```

```python
# Minimum Dry Bulb temperature observed in January (in degrees Celsius)
weather['2019-01']['Dry Bulb Temperature C'].min()
```

```python
# Maximum Dry Bulb temperature observed during first two weeks of February (in degrees Celsius)
weather['2019-02']['Dry Bulb Temperature C'].max()
```

```python
# Minimum Dry Bulb temperature observed during frist two weeks of February (in degrees Celsius)
weather['2019-02']['Dry Bulb Temperature C'].min()
```

```python
# The maximum temperatures observed in January and the first part of February were both equal to 16.7C, 
# however, the minimum temperature was lower in February - 23.3C compared to -20.0C in January.
```

```python
# Daily average 'Dry Bulb' tremperatures in Celsius January 1, 2019 - February14, 2019 
weather.groupby(weather.index.date)['Dry Bulb Temperature C'].mean()
```

```python
# Daily average 'Dry Bulb' tremperatures in Celsius January 1, 2019 - February14, 2019 
weather.groupby(weather.index.month)['Dry Bulb Temperature C'].std()
```

```python
# The above output shows that daily average temperatures fluctuated significantly (from about -16.3 C on February 7 to  about + 8.3 C on February 3),
# with the standard deviation in January of about 6.5, and 8.4 in February. 
```

```python
# Daily mean dry bulb temperatures at or below freezing (<= 0 degrees Celsius) 
mean_below_freezing = weather.groupby(weather.index.date)['Dry Bulb Temperature C'].mean() <= 0
#number of the days when daily mean temperature remained at or below freezing
mean_below_freezing.sum()
```

```python
# There were 20 days during this period when daily mean temperatures did not rise above freezing. 
```

```python
# Average wet bulb temperature (reflecting saturation) in C for each day 
weather.groupby(weather.index.date)['Wet Bulb Temperature C'].mean()
```

```python
# Difference between daily average 'Dry Bulb Temperature' and daily average 'Wet Bulb Temperature'
dry_wet_difference =weather.groupby(weather.index.date)['Dry Bulb Temperature C'].mean() - weather.groupby(weather.index.date)['Wet Bulb Temperature C'].mean()
dry_wet_difference
```

```python
# min difference in degrees Celsius
dry_wet_difference.min()
```

```python
# max difference in degrees Celsius
dry_wet_difference.max()
```

```python
# Daily mean 'Wet bulb' temperatures (Celsius), as the temperatures air cools to when water is added, 
# were from 0.4 to 5.8 degrees lower than 'Dry bulb" temperatures. 
```
