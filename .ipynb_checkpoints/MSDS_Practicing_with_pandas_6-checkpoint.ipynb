```python
# Define functions for analyzing missing values
def count_missing(vec):
    """Counts the number of missing values in a vector
    """
    #Boolean, vector of missing values
    null_vec=pd.isnull(vec)
    #sum the null vector (False=0)
    null_count=np.sum(null_vec)
    #return the number of missing values
    return null_count
```

```python
def prop_missing(vec):
    """Calculates percentage of missing values in a vector
    """
    #number of missing values
    num=count_missing(vec)
    #total number of values in the vector
    dem=vec.size
    #return percentage of missing values
    return num/dem
    
```

```python
def prop_complete(vec):
    """calculates percentage of non-missing values
    """
    #complete = 1 - missing
    return 1 - prop_missing(vec)
```

```python
#Count missing values in the data frame
number_missing = adults.apply(count_missing)
#Display results
number_missing
```

```python
#Find the percentage of missing values
percentage_missing=adults.apply(prop_missing)
percentage_missing
```

```python
#Find percentage of non-missing values
percentage_nonmissing = adults.apply(prop_complete)
percentage_nonmissing
```

There are no missing values in numeric columns as zeros in the 'CapitalGain' and 'CapitalLoss' columns are valid answers, and there are no unusual numbers in the rest of the numeric columns.  However, string type columns contain a relatively high number of missing values - about 5.6% for the 'Workclass' and 'Occupation' columns (1836 and 1843 values missing respectively), as well as 583 values missing in the 'NativeCountry' column.

Since there is no one value similar to some summary statistic (e.g., mean or median for numeric fields) that could be used to impute string values, and in order not to decrease the size of the data set, I suggest adding additional categories to the "Workclass", "Occupation",  and "NativeCountry" columns. Both "Workclass" and "Occupation" columns already have categories for "other", so missing values can be marked as "Unknown" category. As for the "NativeCountry" column, it does not have any categories for other countries and the list is not exhaustive. So, we can add the "Other" category.

```python
#Fill missing values in the "Workclass" and "Occupation columns" with Unknown category
adults[['Workclass', 'Occupation']] = adults[['Workclass', 'Occupation']].apply(lambda x: x.fillna(value="Unknown"))
```

```python
#Replace missing values in the "NativeCountry" column with "Other"
adults['NativeCountry'].fillna(value="Other", inplace=True)
```
